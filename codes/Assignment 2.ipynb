```python
from math import factorial as fac
from numpy import *
from array import *
import matplotlib.pyplot as plt
```

```python
print("Let the number of terms are n" )
print("Calculating the exponential of x" )

x=1

n = 100
```

```python
def function(x,n):
    product = 1
    result = 1
    for i in range(1,n+1):
        product = product * x/(i)
        result = result + product
    return result
```

```python
function(x,n)
```

```python
print(exp(x))
```

```python
m = linspace(0,100,1000)
y = empty(len(m),dtype="object")
print(y)
```

```python
for i in range(len(m)):
    y[i] = function(m[i],n)
    
print(y)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
